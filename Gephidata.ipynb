```python
import pandas as pd
import re
```

```python
df = pd.read_csv('tweets.csv')
```

```python
df.head()
```

```python
# Function to extract mentions from a tweet
def extract_mentions(tweet):
    if isinstance(tweet, str):
        return re.findall(r'@(\w+)', tweet)
    return []

# Create an edge list from the mentions
edges = []
for _, row in df.iterrows():
    source = row['username']
    mentions = extract_mentions(row['tweets'])
    for target in mentions:
        edges.append({'source': source, 'target': target})
```

```python
# Convert edges to a DataFrame
edges_df = pd.DataFrame(edges)
```

```python
edges_df.head(20)
```

```python
# Save the edge list for Gephi
edge_file_path = 'gephi_edges.csv'
edges_df.to_csv(edge_file_path, index=False)
```
